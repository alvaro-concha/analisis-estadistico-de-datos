## Ejercicio 1
Considerar una variable aleatoria $x$ que sigue una distribución normal con parámetro $\mu$ desconocido y $\sigma = 5.9$.
En un experimento se miden los siguiente valores de $x: (67.6, 57.4, 63.0, 68.0, 63.1)$.

Estimar el valor del parámetro $\mu$ y su intervalo con un nivel de confianza del 90%.

## Solución
Sea la variable aleatoria $X \sim \mathcal{N}(\mu, \sigma^2)$, con desviación estándar conocida.

Sean $n$ observaciones $\vec{x} = (x_1, \dots, x_n)$ de la variable aleatoria $X$.

Luego, el intervalo de confianza de $CL\%$ para la media es
$$
\hat{\mu} = \bar{x} \pm q \frac{\sigma}{\sqrt{n}},
$$
siendo $\bar{x} \sim \mathcal{N}(\mu, \sigma^2/n)$ el promedio de la muestra y $q$ es tal que
$$
\int_{\mu - q\sigma}^{\mu + q\sigma} f(x) dx = CL\%.
$$

En el caso en que el intervalo de confianza sea del 90%, $q=1.64$.

Por lo tanto, en este ejercicio, el intervalo es
$$
\hat{\mu} = 63.8 \pm 4.3.
$$

## Ejercicio 3
Simular una muestra de dos variables normales estándar.

Calcular el intervalo de confianza $1\sigma$ del parámetro $\mu$ asumiendo que la desviación estándar $\sigma = 1$ es conocida.

Verificar si el parámetro $\mu$ está contenido dentro del intervalo.

Repetir la simulación 1000 veces para estimar la probabilidad de cobertura del intervalo.

Comparar la probabilidad de cobertura con el nivel de confianza.

## Solución
Sean las $n$ variables aleatorias iid $X_i \sim \mathcal{N}(\mu, \sigma^2)$ con distribución normal estándar.

Suponemos que el parámetro $\sigma=1$ es conocido.

Para cada realización del experimento, vamos a estimar el intervalo de confianza $1\sigma$ de la media como
$$
\hat{\mu} = \bar{x} \pm q \frac{\sigma}{\sqrt{n}},
$$
con $q=1$ y vamos a verificar si el valor verdadero $\mu=0$ está contenido en él, es decir si
$$
\mu \in \left[\bar{x} - q \frac{\sigma}{\sqrt{n}}, \bar{x} + q \frac{\sigma}{\sqrt{n}}\right].
$$

Repetimos la simulación $n_{\mathrm{realizaciones}} = 1000$ veces.

Estimamos la probabilidad de cobertura como
$$
P_{\mathrm{cobertura}} = \frac{n_{\mathrm{éxitos}}}{n_{\mathrm{realizaciones}}}.
$$

Esta probabilidad es igual al nivel de confianza del intervalo $1\sigma$, esto es
$$
P_{\mathrm{cobertura}} = 68\%.
$$

## Ejercicio 4
En un experimento se miden los valores $(13.4, 8.52, 12.7, 9.9, 12.8)$.

Asumiendo que los datos siguen una distribución normal con parámetros $\mu$ y $\sigma$ desconocidos, calcular el intervalo de Student al 90% de nivel de confianza en términos de la media muestral $\bar{x}$ y la desviación estándar muestral $s$.

## Solución
Sean $n$ realizaciones
$$
\mathbf{x} = (x_1, \dots, x_n),
$$
de una variable aleatoria con distribución normal
$$
X \sim \mathcal{N}(\mu, \sigma^2),
$$
con media y varianza desconocidas.

Luego la variable aleatoria 
$$
t = \frac{\bar{x} - \mu}{s / \sqrt{n}},
$$
donde $\bar{x}$ y $s$ son la media y la desviación estándar muestral, tiene distribución t de Student
$$
T \sim \mathrm{Student}(\nu),
$$
con grados de libertad
$$
\nu = n - 1.
$$

La distribución t de Student se aproxima a una distribución normal estándar cuando los grados de libertad tienden a infinito, es decir
$$
\mathrm{Student}(\nu) \xrightarrow[\nu \to \infty]{} \mathcal{N}(0, 1).
$$

En el caso de una variable gaussiana con media y varianza desconocidas, podemos estimar el intervalo de confianza (intervalo de Student) de 90% de la media, como
$$
\hat{\mu} = \bar{x} \pm q \frac{s}{\sqrt{n}},
$$
con $q=1.64$.

Y, para este ejercicio, resulta
$$
\hat{\mu} = 11.5 \pm 1.6.
$$

## Ejercicio 5
Simular dos variables normales $x_1$ y $x_2$ con parámetros $\mu_1 = 10.7$, $\mu_2 = 8.3$, $\sigma_1 = 1.7$, $\sigma_2 = 2.4$, y correlación $\rho = 0.78$.

Considerar la elipse de $95\%$ de confianza del parámetro $\mu = (\mu_1 , \mu_2)$.

Simular varios valores de $x_1$ y $x_2$ y verificar si las elipses correspondientes contienen a $\mu$.

Repetir la simulación 10000 veces para estimar la probabilidad de cobertura y comparar con el nivel de confianza.

## Solución
Sea la variable aleatoria binormal
$$
X = (X_1, X_2) \sim \mathcal{N}_2(\boldsymbol{\mu}, \Sigma),
$$
con media
$$
\boldsymbol{\mu} = (\mu_1, \mu_2),
$$
y matriz de covarianza
$$
\Sigma = \begin{pmatrix}
\sigma_1^2 & \rho\sigma_1\sigma_2\\
\rho\sigma_1\sigma_2 & \sigma_2^2
\end{pmatrix}.
$$

Su función de costo likelihood es
$$
q(\mathbf{X}) = (\mathbf{X}-\boldsymbol{\mu})^T \Sigma^{-1} (\mathbf{X}-\boldsymbol{\mu}) \sim \chi_2^2.
$$

Las regiones de probabilidad $r\sigma$ tienen como frontera a las elipses que satisfacen
$$
q(x_1, x_2) = r^2.
$$

Conociendo la distribución de probabilidad $\chi_2^2$
$$
P[q(x_1, x_2) = r^2] = \frac{1}{2} \exp\left(-\frac{r^2}{2}\right),
$$
la probabilidad acumulada dentro de la región $r\sigma$ es
$$
P[q(x_1, x_2) \le r^2] = 1 - \exp\left(-\frac{r^2}{2}\right).
$$

Por lo que para encerrar una probabilidad total $CL\%$ se debe utilizar una elipse de tamaño
$$
r = \sqrt{-2\ln\left( 1 - \frac{CL}{100} \right)}.
$$

## Ejercicio 6 (opcional)
Considerar la función de costo,
$$
J(S, \beta) = 2 \sum_{i=1}^n \left\{ \mu_i(S, \beta) - k_i - k_i \log\left[\mu_i(S, \beta)/k_i \right] \right\},
$$
con
$$
\mu_i(S, \beta) = S r_{i}^{-\beta},
$$
donde $S$ y $\beta$ son dos parámetros desconocidos.

Los datos del experimento son el número de partículas $k_i$ medidas por un detector ubicado a una distancia $r_i$,

| Detector | $r_i$ | $k_i$ |
|:-:|:-:|:-:|
| 1 | 0.764 | 33 |
| 2 | 1.052 | 19 |
| 3 | 1.236 | 11 |
 
Dibujar la función de costo como función de $S$ y $\beta$.

Con una minimización numérica encontrar los estimadores de $S$ y $\beta$.

Dibujar la región de confianza de $1\sigma$ correspondiente a $J = J_{\min} + 1$, con $J_{\min}$ el
mínimo de la función de costo.

## Solución
En general, tenemos que el intervalo de likelihood, definido como la región donde la función de costo es menor a cierto umbral $q^2$
$$
J(\boldsymbol{\theta}) - J_{\min} \le q^2,
$$
es igual al intervalo de confianza $q\sigma$ de la distribución del estimador del parámetro $\boldsymbol{\theta}$.

Si conocemos la forma funcional de $J(\boldsymbol{\theta})$, entonces podemos calcular directamente el intervalo de confianza a partir de ella.

Minimizando numéricamente, se obtuvieron los estimadores
$$
S_{MLE} = 19.12,
$$
y
$$
\beta_{MLE} = 2.09,
$$
valores donde $J(\boldsymbol{\theta})$ adopta el valor
$$
J_{\min} = 0.33.
$$

Se observa que para valores pequeños de $q$, como
$$
q \lesssim 1,
$$
las superficies de nivel definidas por
$$
J(\boldsymbol{\theta}) - J_{\min} = q^2,
$$
pueden aproximarse mediante elipses.

Mientras que, si,
$$
q \gg 1,
$$
las superficies de nivel difieren mucho de elipses.

Nota: para estimar los errores de los estimadores $\boldsymbol{\theta}$ podría usarse la función ``scipy.minimize`` sobre la función de costo, como vimos en clase.

## Ejercicio 7 (Para entregar)
Considerar $n=32$ lanzamientos de una moneda cargada con probabilidad $p$ de salir cara.

El número de caras $K$ es una variable aleatoria con distribución binomial
$$
K \sim \mathrm{Binomial}(n, p).
$$

Considerar una tanda de lanzamientos en el que salen $k$ caras.

El estimador del parámetro $p$ es $\hat{p} = k/n$.

El intervalo de confianza estándar de $p$ con un nivel de confianza del $95\%$ es
$$
\hat{p} \pm \underbrace{1.96}_{=q} \sqrt{\frac{\hat{p} (1 -\hat{p})}{n}},
$$
donde $q = 1.96$ es el cuantil normal estándar para un $95\%$ CL.

Simular 10000 tandas de lanzamientos asumiendo $p = 0.2$.

Verificar para cada tanda si el intervalo incluye o no a $p = 0.2$.

Estimar la cobertura del intervalo.

Decidir si el intervalo es exacto y comparar la cobertura con respecto al nivel de confianza.

Opcional: Barrer $p$ en el intervalo $[0,1]$, graficar y ver si hay valores de $p$ para los cuales la cobertura es particularmente mala.

## Solución
Sea la variable aleatoria binomial
$$
K \sim \mathrm{Binomial}(n, p)
$$
con $n=32$ conocido y $p$ fijo desconocido.

Ante una realización de la variable aleatoria
$$
K=k,
$$
el estimador de maximum likelihood del parámetro $p$, conociendo $n$, es
$$
\hat{p} = \frac{k}{n}.
$$

La función de costo asociada es
$$
J(p) = 2 n \left[ \hat{p}\ln\left( \frac{\hat{p}}{p} \right) + (1 - \hat{p})\ln\left( \frac{1-\hat{p}}{1-p} \right) \right],
$$
con la cual podemos obtener un intervalo de likelihood exacto, encontrando los puntos que satisfacen
$$
J(p) = q^2,
$$
y este intervalo es igual al intervalo de confianza $q\sigma$.

Además, podemos aproximar la función de costo alrededor de $p = \hat{p}$ como una parábola, obteniendo el intervalo de confianza aproximado para el parámetro $p$
$$
\hat{p} \pm q \sqrt{\frac{\hat{p} (1 -\hat{p})}{n}},
$$
donde $q=1.64$ corresponde a un nivel de confianza normal estándar del $95\%$.

Esta aproximación equivale a tratar la distribución binomial como si fuera gaussiana.

La cual es una buena aproximación si $n\gg 1$ y si $p$ está alejado de los valores extremos $p=0$ y $p=1$.

Pues el estimador de maximum likelihood tiene una distribución
$$
\begin{align*}
\hat{p} &= \frac{k}{n}\\
&\sim \frac{1}{n} \mathrm{Binomial}(n, p)
\end{align*}
$$

Por último, vemos que los intervalos realizados a partir de un estimador corregido
$$
\hat{p} = \frac{k+2}{n+4},
$$
tienen una cobertura más cercana al nivel de confianza, en configuraciones de parámetros $(n, p)$ donde el estimador de maximum likelihood falla.
